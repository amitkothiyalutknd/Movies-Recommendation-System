```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    nameList = []
    for i in ast.literal_eval(obj):
        nameList.append(i['name'])
    return nameList        
```

```python
Movies['genres'] = Movies['genres'].apply(convert)
```

```python
Movies['keywords'] = Movies['keywords'].apply(convert)
```

```python
def firstThreeNames(obj):
    starName = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            starName.append(i['character'])
            counter += 1
        else:
            break
    return starName        
```

```python
Credits['cast'].apply(firstThreeNames)
```

```python
Credits['cast'] = Credits['cast'].apply(firstThreeNames)
```

```python
def fetchDirector(obj):
    dirName = []
    counter = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            dirName.append(i['name'])
            break
    return dirName
```

```python
Credits['crew'].apply(fetchDirector)
```

```python
Credits['crew'] = Credits['crew'].apply(fetchDirector)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
 def stem(text):
        y = []
        for i in text.split():
            y.append(ps.stem(i))
        return " ".join(y)
```

```python
def recommend(movie):
    movieIndex = newMoviesDF[newMoviesDF['title'] == movie].index[0]
    distances = similarity[movieIndex]
    moviesList = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[:5]    
    for i in moviesList:
        print(newMoviesDF.iloc[i[0]].title)
```
